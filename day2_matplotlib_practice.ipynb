```python
import matplotlib.pyplot as plt
plt.figure(figsize=(12, 4))
expenses = [45000, 52000, 48000, 61000, 55000, 49000]
months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun"]
plt.subplot(1,2,1)
plt.plot(months,expenses,color='red')
plt.title("Monthly Expenses H1 2026")
plt.xlabel("Month")
plt.ylabel("Expense (₹)")
graph_colors=[]
for expense in expenses:
    if expense > 50000:
        graph_colors.append('red')
    else:
        graph_colors.append('green')
plt.subplot(1,2,2)
plt.bar(months,expenses,color=graph_colors)
plt.title("Monthly Expenses H1 2026")
plt.xlabel("Month")
plt.ylabel("Expense (₹)")
plt.show()


user_input=0
total_amount=0
while True: 
    user_input = input("Enter numbers to be added, if you have entered all numbers then type done: ")
    if not user_input.isdigit() and user_input != 'done':
        print("Invalid number entered, please enter a valid number")
        continue
    elif user_input == 'done':
        print(f'The total amount is {total_amount}')
        break
    total_amount=total_amount+ int(user_input)
    
```
